```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tuan.csv').drop('Unnamed: 0', axis=1)
df
```

```python
dic = {}
for i in df.values:
    print(i[1], end=':')
    dic[i[1]] = input()
```

```python
for i, j in dic.items():
    df.loc[lambda df: df.Name == i, 'Age'] = j
```
